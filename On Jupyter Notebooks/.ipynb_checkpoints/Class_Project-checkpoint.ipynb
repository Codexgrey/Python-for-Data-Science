### Grouping by Year

```python
# grouping by successful and failed transcations per month
gps_year = pd.DataFrame(data[data["status"] == "successful"].groupby(["txnYear", "txnMonth", "status"])["no of txns"].count().reset_index())
gpf_year = pd.DataFrame(data[data["status"] == "failed"].groupby(["txnYear", "txnMonth", "status"])["no of txns"].count().reset_index())
```

```python
gps_year
```

```python
gpf_year
```

### Grouping by Month

```python
# grouping by successful transcations per month for 2019 and 2020
gps_19 = gps_year[gps_year["txnYear"] == 2019]
gps_20 =  gps_year[gps_year["txnYear"] == 2020]

# grouping failed transcations per month for 2019 and 2020
gpf_19 = gpf_year[gps_year["txnYear"] == 2019]
gpf_20 =  gpf_year[gps_year["txnYear"] == 2020]

print(f"""
    {gps_19}+\n, 
    {gps_20}+\n,
    {gpf_19}+\n, 
    {gpf_20}+\n
""")
```

##### Grouping by Month names

```python
# change "txnMonth" series type to string
gps_19["txnMonth"] = gps_19["txnMonth"].astype("str")
gps_20["txnMonth"] = gps_20["txnMonth"].astype("str")
gpf_19["txnMonth"] = gpf_19["txnMonth"].astype("str")
gpf_20["txnMonth"] = gpf_20["txnMonth"].astype("str")

# using datetime to get month names instead of as numbers
gps_19["txnMonth"] = gps_19["txnMonth"].apply(lambda c : dt.strptime(c, "%m"))
gps_20["txnMonth"] = gps_20["txnMonth"].apply(lambda c : dt.strptime(c, "%m"))
gpf_19["txnMonth"] = gpf_19["txnMonth"].apply(lambda c : dt.strptime(c, "%m"))
gpf_20["txnMonth"] = gpf_20["txnMonth"].apply(lambda c : dt.strptime(c, "%m"))

## sorting
gps_19.sort_values( by = "txnMonth", inplace = True)
gps_20.sort_values( by = "txnMonth", inplace = True)
gpf_19.sort_values( by = "txnMonth", inplace = True)
gpf_20.sort_values( by = "txnMonth", inplace = True)

## getting months by names
gps_19["txnMonth"] = gps_19["txnMonth"].apply(lambda c : c.strftime("%b"))
gps_20["txnMonth"] = gps_20["txnMonth"].apply(lambda c : c.strftime("%b"))
gpf_19["txnMonth"] = gpf_19["txnMonth"].apply(lambda c : c.strftime("%b"))
gpf_20["txnMonth"] = gpf_20["txnMonth"].apply(lambda c : c.strftime("%b"))

# print data
print(f"""
    {gps_19} +\n, 
    {gps_20} +\n,
    {gpf_19} +\n, 
    {gpf_20} +\n
""")
```

```python
# plotting a bar chart comparing per year; monthly successful transations
plt.figure(figsize = (12, 8), facecolor="azure")
ind = np.arange(11)

plt.bar(gps_20["txnMonth"], gps_20["no of txns"], width=0.5, color="purple", label="2020")
plt.bar(gps_19["txnMonth"], gps_19["no of txns"], width=0.5, color="seagreen", label="2019")


plt.title("Monthly Successful Transactions Trend",fontsize = 20, pad = 15)
plt.xlabel("Month", size=15)
plt.ylabel("Transaction Volume", size=15)
plt.legend(loc="best", shadow=True)

plt.show()
```

```python
# plotting a bar chart comparing per year; monthly failed transations
plt.figure(figsize = (12, 8), facecolor="azure")

plt.bar(gpf_20["txnMonth"], gpf_20["no of txns"], width=0.5, color = "purple", label="2020")
plt.bar(gpf_19["txnMonth"], gpf_19["no of txns"], width=0.5, color = "seagreen", label="2019")


plt.title("Monthly Failed Transactions Trend", fontsize=20, pad=15)
plt.xlabel("Month", size=15)
plt.ylabel("Transaction Volume", size=15)
plt.legend(loc="best", shadow=True)

plt.show()
```
